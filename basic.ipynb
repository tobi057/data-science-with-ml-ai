```python
### pie
```

```python
x =np.array([1,2,3,4])
labels = ['a','b','c','d']
plt.pie(x, labels = labels)
plt.show()
```

```python
colour =['red','green','blue','yellow']
myexplode = [0,0.2,0,0]
x =np.array([1,2,3,4])
labels = ['a','b','c','d']
plt.pie(x, labels = labels, explode = myexplode, shadow = True,autopct ='%1.1f%%',colors=colour)
plt.legend('my pie chart')
plt.show()
```

```python
x = np.array([1,2,3,4,5])
colour = ['red','green','blue','yellow','brown']
labels = ['a','b','c','d','e']
myexplode = [0,0,0.2,0,0]
plt.pie(x,colors=colour, labels=labels, explode=myexplode, shadow=True, autopct='%1.1f%%')
plt.legend('my pie chart')
plt.show()
```

```python
colour =['red','green','blue','yellow']
myexplode = [0,0.2,0,0]
x =np.array([1,2,3,4])
labels = ['a','b','c','d']
plt.pie(x, labels = labels, explode = myexplode, shadow = True,autopct ='%1.1f%%',colors=colour)
plt.legend('my pie chart')
plt.title('my pie chart')
plt.show()
```
